```python
from distributions import levy
import torch
import seaborn as sns
import pandas as pd
```

```python
data = dict()
data['x'] = []
data['y'] = []

for i in range(50):
    point = levy(2, 3)
    data['x'].append(float(point[0]))
    data['y'].append(float(point[1]))

df = pd.DataFrame(data=data)

```

```python
print(df)
```

```python
ax = sns.relplot(x="x", y="y", data=df, kind='line')
```
